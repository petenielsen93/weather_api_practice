```python
# Import the libraries.
import requests
import json
import gmaps
import pandas as pd
# Import the API keys.
from config import weather_api_key
from config import g_key
```

```python
top_50 = pd.read_csv("top50.csv")
top_50.head()
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=" + weather_api_key
cities = top_50["City"]



print(cities)
```

```python
city_data = []

for city in cities:
    city_url = url + "&q=" + city.replace(" ","+")
    city_weather = requests.get(city_url).json()
    city_lat = city_weather["coord"]["lat"]
    city_lng = city_weather["coord"]["lon"]
    city_max_temp = city_weather["main"]["temp_max"]
    city_humidity = city_weather["main"]["humidity"]
    city_desc = city_weather["weather"][0]["description"]
    city_data.append({"Lat":city_lat,
                     "Lng": city_lng,
                     "Max Temp": city_max_temp,
                     "Humidity": city_humidity,
                     "Description": city_desc})
    
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df
```

```python
cities_df = pd.concat([top_50, city_data_df], axis = 1, ignore_index=True)
cities_df.columns=["city","state","lat","lng","temp","humidity","desc"]
cities_df
```

```python
info_box_template = """
<dl>
<dt>City</dt><dd>{city}</dd>
<dt>Current Weather</dt><dd>{desc} and {temp} Â°F</dd>
</dl>
"""

locations = cities_df[["lat","lng"]]

weather_info = [info_box_template.format(**row) for index, row in cities_df.iterrows()]
```

```python
marker_layer = gmaps.marker_layer(locations, info_box_content=weather_info)

fig = gmaps.figure(center=(100.0, 100.0), zoom_level=1.5)

fig.add_layer(marker_layer)
```
