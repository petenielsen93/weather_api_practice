```python
# Import the requests library.
import requests
import json
# Import the API key.
from config import weather_api_key
import gmaps
from config import g_key
```

```python
#Starting URL for Weather Map API Call. Calling by ZIP Code 32751

url = "http://api.openweathermap.org/data/2.5/weather?zip=32751&appid=" + weather_api_key

print(url)
```

```python
#API call based on the 25 cities surrounding the lat/lon for Maitland

cfl_cities = "http://api.openweathermap.org/data/2.5/find?lat=28.5383&lon=-81.3792&cnt=25&appid=" + weather_api_key
print(cfl_cities)
```

```python
#requesting direct download of data for central fl cities based on API link above. "requests" uses our requests library,
#and combines with get(), which retrieves the info from the source. 
#Using .json() transforms the pulled data in JSON format
central_fl = requests.get(cfl_cities)
central_json = central_fl.json()
```

```python
#open entire JSON file
central_fl.json()
```

```python
#get the JSON data
norlando = central_fl.json()
```

```python
norlando["list"]
```

```python
norlando["list"][0]["name"]
```

```python
#loop through list objects in norlando to pull the names from JSON object and create list of city names 

names = []

for i in norlando["list"]:
    names.append(i["name"])
print(names)
```

```python
#loop for lats of norlando cities 
lats = []

for i in norlando["list"]:
    lats.append(i["coord"]["lat"])
print(lats)
```

```python
#loop for lons of norlando cities 
lons = []

for i in norlando["list"]:
    lons.append(i["coord"]["lon"])
print(lons)
```

```python
#loop for current temp
temp = []

for i in norlando["list"]:
    temp.append(i["main"]["temp"])
print(temp)
```

```python
#loop for current humidity
humid = []

for i in norlando["list"]:
    humid.append(i["main"]["humidity"])
print(humid)
```

```python
#loop for current rain
rain = []

for i in norlando["list"]:
    rain.append(i["rain"])
print(rain)
```

```python
#loop for current weather status
status = []

for i in norlando["list"]:
    status.append(i["weather"][0]["main"])
print(status)
```

```python
#loop for current weather description
desc = []

for i in norlando["list"]:
    desc.append(i["weather"][0]["description"])
print(desc)
```

```python
#convert kelvin to fahrenheit
temp_temp = list(map(float, temp))
temp_temp
```

```python
temp_in_F = []
for x in temp_temp:
    Ftemp =round((x - 273.15) * 9/5 + 32,2)
    temp_in_F.append(Ftemp)


```

```python
#create dataframe of pulled data
import pandas as pd
norlando_df = pd.DataFrame({"city_name":names, "latitude":lats, "longitude":lons, "temperature (F)":temp_in_F, "humidity_pct":humid, "rain":rain, "weather_status":status, "weather_desc":desc})
norlando_df
#if applied to weather api map, could act as a live weather map for heroku app???
#national map possible with several api calls to major metro areas???
```

```python
info_box_template = """
<dl>
<dt>City</dt><dd>{city_name}</dd>
<dt>Current Weather</dt><dd>{weather_status} and {temperature (F)} F</dd>
</dl>
"""

locations = norlando_df[["latitude","longitude"]]

weather_info = [info_box_template.format(**row) for index, row in norlando_df.iterrows()]
```

```python
marker_layer = gmaps.marker_layer(locations, info_box_content=weather_info)

fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

fig.add_layer(marker_layer)
```
